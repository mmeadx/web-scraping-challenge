```python
#Importing Dependencies
import pandas as pd
from bs4 import BeautifulSoup
import requests
from splinter import Browser
```

```python
# URL for NASA Mars News
nasa_url = "https://mars.nasa.gov/news/?page=0&per_page=40&order=publish_date+desc%2Ccreated_at+desc&search=&category=19%2C165%2C184%2C204&blank_scope=Latest"
```

```python
response = requests.get(nasa_url)
```

```python
#Check Response
response
```

```python
#Make Soup
soup = BeautifulSoup(response.text, 'html.parser')
soup
```

```python
# Grab first News Headline
news_headline = soup.find('div', class_="content_title").text
print(news_headline)
```

```python
# Grab first description
news_p = soup.find('div', class_="rollover_description_inner").text
print(news_p)
```

```python
executable_path = {'executable_path': '/usr/local/bin/chromedriver'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Get browser to go to JPL website
jpl_url = 'https://www.jpl.nasa.gov/spaceimages/?search=&category=Mars'
browser.visit(jpl_url)
```

```python
# Make Soup
html = browser.html
soup = BeautifulSoup(html, 'html.parser')

soup
```

```python
# Get Featured Image URL
image = soup.find('a', class_="button fancybox")['data-fancybox-href']
featured_image_url = "https://www.jpl.nasa.gov" + image
featured_image_url
```
